```python
data = pd.read_csv("Groceries_dataset.csv")
products = data['itemDescription'].unique()

#one hot encoding the products:

dummy = pd.get_dummies(data['itemDescription'])
data.drop(['itemDescription'], inplace =True, axis=1)

data = data.join(dummy)

data.head()
```

```python
# Transaction: If a customer bought multiple products in one day, it will be considered as 1 transaction:

data1 = data.groupby(['Member_number', 'Date'])[products[:]].sum()
data1 = data1.reset_index()[products]

print("New Dimension", data1.shape)
data1.head()
```

```python
# Ensure data1 is in a numerical format before applying apriori
data1 = data.groupby(['Member_number', 'Date'])[products[:]].sum()
data1 = data1.reset_index()[products]

# We will not replace non-zero values with product names here
# as apriori requires numerical data
print("New Dimension", data1.shape)
data1.head()

```

```python
print("Total Number of Transactions:", len(data1))
```

```python
#Removing Zeros, Extracting the list of items bought per customer

x = data1.values
x = [sub[~(sub==0)].tolist() for sub in x if sub [sub != 0].tolist()]
transactions = x
transactions[0:10]
```

```python
# Convert the dataframe to a list of lists for apriori
transactions_list = []
for index, row in data1.iterrows():
    transaction = list(row[row > 0].index)  # Get column names where value is > 0
    if transaction:
        transactions_list.append(transaction)

# Removing Zeros, Extracting the list of items bought per customer
# The previous approach of converting to numpy array and then to list resulted in string issues.
# This approach directly creates a list of lists from the numerical dataframe.
# x = data1.values
# x = [sub[~(sub==0)].tolist() for sub in x if sub [sub != 0].tolist()]
transactions = transactions_list

rules = apriori(transactions, min_support = 0.0001, min_confidence = 0.05, min_lift = 1.0,
max_length = 2, target = "rules")
association_results = list(rules)
print(association_results[0])
```

```python
for item in association_results:
    # Access items from the frozenset
    pair = item.items
    items = [x for x in pair]

    # Check if there are at least two items in the rule
    if len(items) >= 2:
        # Access ordered statistics
        ordered_stats = item.ordered_statistics
        if ordered_stats:
            # Check if the first ordered statistic has enough elements
            if len(ordered_stats[0]) >= 4:
                print("Rule : ", items[0], " -> " + items[1])
                print("Support : ", str(item.support))
                print("Confidence : ",str(ordered_stats[0].confidence))
                print("Lift : ", str(ordered_stats[0].lift))
                print("=============================")
            else:
                print(f"Skipping rule with insufficient ordered statistics: {items}")
        else:
            print(f"Skipping rule with no ordered statistics: {items}")
    else:
        print(f"Skipping rule with less than two items: {items}")
```
